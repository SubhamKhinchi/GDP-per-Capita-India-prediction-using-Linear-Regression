```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("GDP_per_Capita_India.csv")
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(df.Year, df.GDP_per_Capita)
plt.xlabel("Year")
plt.ylabel("GDP per Capita (in US$)")
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["Year"]],df["GDP_per_Capita"])
```

```python
year = 2020
```

```python
reg.predict([[year]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.figure(figsize=(12,7))
plt.scatter(df.Year, df.GDP_per_Capita)
plt.xlabel("Year")
plt.ylabel("GDP per Capita (in US$)")
plt.plot(df["Year"], reg.predict(df[["Year"]]), color='red')
plt.show()
```

```python
future_years = pd.DataFrame(data={"Years" : [2020,2021,2022,2023,2024]})
```

```python
Predict_GDP = reg.predict(future_years)
```
